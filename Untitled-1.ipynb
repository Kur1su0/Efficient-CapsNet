```python
"""
load mhd data from a path
"""
import numpy as np
import SimpleITK as sitk
import os
import matplotlib.pylab as plt
# ct_scans = sitk.GetArrayFromImage(sitk.ReadImage("training_001_ct.mhd", sitk.sitkFloat32))
import cv2
PATH = "dataset/Testing/Diseased/OAS31119_MR_d0069.mhd"

```

```python

def load_mhd(path):
    """
    load mhd data from a path
    """
    ct_scan = sitk.GetArrayFromImage(sitk.ReadImage(path, sitk.sitkFloat32))

    return ct_scan
def plot_mhd(ct_scan):
    
    plt.imshow(ct_scan)
    #resize the image
    resized_image = cv2.resize(ct_scan, (28,28))
    plt.imshow(resized_image)
    # use plt.savefig(...) here if you want to save the images as .jpg, e.g.,

#def main
if __name__ == "__main__":
    ret = load_mhd(PATH)
    print(ret.shape)
    plot_mhd(ret)
    
plt.show()
```
